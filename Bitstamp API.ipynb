```python
import requests
import pandas as pd
import datetime as dt
```

```python
start = "2011-08-01"
end = dt.datetime.today()
dates = pd.date_range(start, end)
dates = [int(x.value/(10**9)) for x in dates]
df = pd.DataFrame(index = dates, columns =  [["close", "high", "low", "open", "timestamp", "volume"]])
```

```python
def request_bitstamp_data(pair="btcusd", step=86400, limit=1000):
    """
    Request bitstamp API to get OHLC data 
    and write it a list of dictionaries
    """
    data = []
    count = len(dates) 
    for first, last in zip(dates, dates[1:]):
        parameters = {"step": step,
                      "limit": limit,
                      "start": first,
                      "end" : last
                      }
        count -= 1
        print(f"Bars to download left: {count}")
        req = requests.get(f"https://www.bitstamp.net/api/v2/ohlc/{pair}/", params = parameters)
        req = req.json()["data"]["ohlc"]
        data += req
        
    print(f"\nData download for pair: {pair.upper()} completed")
    return data
```

```python
# Request and write ohlc data
data = request_bitstamp_data()

## Write data to dateframe and export csv
df = pd.DataFrame(data)
df = df.drop_duplicates()
df = df.reset_index()
df = df.drop(labels = ["index"], axis = 1)
df.to_csv("btcusd_data.csv", sep=",", columns = ["close", "high", "low", "open", "timestamp", "volume"])
```
